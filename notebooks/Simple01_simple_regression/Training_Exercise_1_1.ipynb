```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Datos extraídos directamente del ejercicio 1.1
data = {
    'Age': [45, 28, 65, 48, 35, 30, 75, 30, 50, 27, 65, 30, 28,
            20, 35, 40, 25, 30, 60, 45, 25, 40, 25, 35, 35, 40],
    'Expenditures': [150, 300, 100, 150, 400, 300, 50, 350, 200, 250, 150, 350, 320,
                    200, 450, 250, 300, 300, 120, 180, 250, 200, 280, 380, 420, 220]
}

df = pd.DataFrame(data)

# --- CÁLCULOS ---
# (d) Media global
mean_all = df['Expenditures'].mean()

# (e) Medias segmentadas (Corte en 40 años)
mean_old = df[df['Age'] >= 40]['Expenditures'].mean()
mean_young = df[df['Age'] < 40]['Expenditures'].mean()

print(f"--- RESULTADOS EJERCICIO 1.1 ---")
print(f"(d) Media de gastos (todos): {mean_all:.2f}")
print(f"(e) Media gastos (Edad >= 40): {mean_old:.2f}")
print(f"    Media gastos (Edad < 40): {mean_young:.2f}")

# --- PREDICCIONES (f) ---
print(f"\n--- PREDICCIONES (f) ---")
print(f"Para un cliente de 50 años: {mean_old:.2f}")
print(f"Para un cliente de 25 años: {mean_young:.2f}")

# --- GRÁFICA ---
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Age', y='Expenditures', hue=(df['Age'] >= 40), palette='coolwarm', s=100)
plt.axvline(40, color='red', linestyle='--', label='Punto de corte (40 años)')
plt.title('Distribución de Gastos por Edad')
plt.grid(True, alpha=0.3)
plt.legend(title='Grupo de Edad')
plt.show()
```
